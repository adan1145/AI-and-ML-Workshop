## Splitting into Train and Test Sets

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,shuffle = True,random_state=100)
```

## Applying our First Model with Scikit-learn: KNN

```python
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
```

### Training our model

### Making Predictions

```python
print("Test set predictions: {}".format(knn.predict(X_test)))
```

### Testing Accuracy

```python
print("Test set accuracy: {:.2f}".format(knn.score(X_test, y_test)))
```

### Improving the Model Performance

Min Max Scaling scales the numeric values of a dataset by transforming them into a range between 0 and 1. The formula used is as follows:

X_std = (X - X.min) / (X.max - X.min)


X_scaled = X_std * (max - min) + min


```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_minmax_scaled = scaler.transform(X_train)
X_test_minmax_scaled = scaler.transform(X_test)
```

```python
knn.fit(X_train_minmax_scaled, y_train)
```

```python
print("Test set accuracy: {:.2f}".format(knn.score(X_test_minmax_scaled, y_test)))
print("Train set accuracy: {:.2f}".format(knn.score(X_train_minmax_scaled, y_train)))
```

### Testing with different values of k

```python
def knn_acc_with_k(X_train,y_train, X_test, y_test,plt):
    train_accuracy = []
    test_accuracy = []
    neighbors = range(1, 11)
    for n in neighbors:
        knn = KNeighborsClassifier(n_neighbors=n)
        knn.fit(X_train, y_train)
        train_accuracy.append(knn.score(X_train, y_train))
        test_accuracy.append(knn.score(X_test, y_test))
    plt.plot(neighbors, train_accuracy, label="train accuracy")
    plt.plot(neighbors, test_accuracy, label="test accuracy")
    plt.ylabel("Accuracy")
    plt.xlabel("neighbors")
    plt.legend()
```

```python
knn_acc_with_k(X_train_minmax_scaled,y_train,X_test_minmax_scaled,y_test,plt)
```

### Plotting results of diffrent values of k

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train_minmax_scaled, y_train)
print(knn.score(X_train_minmax_scaled, y_train))
print(knn.score(X_test_minmax_scaled, y_test))
```

```python
def knn_acc_with_split(X,y,splits,plt,neighbors=5):

    knn = KNeighborsClassifier(n_neighbors = neighbors)

    plt.figure()

    for s in splits:
        scores = []
        for i in range(1,1000):
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1-s,random_state=100)
            knn.fit(X_train, y_train)
            scores.append(knn.score(X_test, y_test))
        plt.plot(s, np.mean(scores), 'bo')

    plt.xlabel('Training set proportion (%)')
    plt.ylabel('accuracy')
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
knn_acc_with_split(X_scaled,y,[0.85,0.8,0.7,0.6,0.5,0.4,0.3],plt)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15,stratify=y,shuffle=True,random_state=100)
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_minmax_scaled = scaler.transform(X_train)
X_test_minmax_scaled = scaler.transform(X_test)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_minmax_scaled, y_train)
print("Test set accuracy: {:.2f}".format(knn.score(X_test, y_test)))
```
